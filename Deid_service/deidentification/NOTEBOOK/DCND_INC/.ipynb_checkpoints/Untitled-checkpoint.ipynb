```python
data = [
('APPOINTMENTVIEW', 1000370, 1119068),
('APPOINTMENTELIGIBILITYINFO', 1, 297342)]
for table_name, start_val, end_val in data:
    obj = Table.objects.filter(metadata__table_name=table_name).last()
    if obj:
        obj.nd_auto_increment_start_value = start_val
        obj.nd_auto_increment_end_value = end_val
        obj.save()
    else:
        print(f"Table not found: {table_name}")
```

```python
from asgiref.sync import sync_to_async
@sync_to_async
def update_table_fast(table_name, start_val, end_val):
    updated = Table.objects.filter(
        metadata__table_name=table_name
    ).update(
        nd_auto_increment_start_value=start_val,
        nd_auto_increment_end_value=end_val
    )

    if updated == 0:
        return f"Table not found: {table_name}"
    return None

```

```python
for table_name, start_val, end_val in data:
    msg = await update_table_fast(table_name, start_val, end_val)
    if msg :
        print(msg)
```

```python
from django.db.models import Count

_db= IncrementalQueue.objects.get(id=22)
grouped_tasks = Task.all_objects.filter(status=0).values('arguments__table_id').annotate(count=Count('id'))
#print(list(counts))

table_ids = [entry['arguments__table_id'] for entry in grouped_tasks if entry['arguments__table_id'] is not None]

table_info = Table.objects.filter(id__in=table_ids, incremental_queue=_db).values('id', 'metadata__table_name')
table_map = {entry['id']: entry['metadata__table_name'] for entry in table_info}

#print(table_info)
#print(table_map)

result = []
for entry in grouped_tasks:
    table_id = entry['arguments__table_id']
    count = entry['count']
    table_name = table_map.get(table_id, 'Unknown')
    result.append({'table_id': table_id, 'table_name': table_name, 'count': count})


total_tasks=0
total_tables = [] 
for r in result:
    if r.get('count',0)>1:
        total_tasks += r.get('count', 0)
        total_tables.append(r.get('table_name'))
        print("Tasks Left: ", r)
print("Total Tasks:", total_tasks)
print("Total Tables:", len(total_tables))
```
